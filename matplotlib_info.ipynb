```python
import matplotlib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
fig = plt.figure()
fig.show()
```

```python
_ = plt.plot([0, 1], [0, 3])
```

```python
plt.plot([0, 1], [0, 3], label='line', color='green')
plt.legend()
plt.xlabel('oś x')
plt.ylabel('oś y')
plt.title('Wykres')
```

```python
plt.plot([0, 1, 2], [0, 3, 2], label='line', color='green')
plt.legend()
plt.xlabel('oś x')
plt.ylabel('oś y')
plt.title('Wykres')
```

```python
x = np.arange(-3, 3, 0.1)
x_sin = np.sin(x)
x_cos = np.cos(x)

plt.plot(x, x_sin)
_ = plt.plot(x, x_cos)
```

```python
plt.subplot(121)
plt.plot([0, 1, 2], [3, 2, 5])

plt.subplot(122)
_ = plt.plot([0, 8, 12], [12, 2, 15], 'r--')
```

```python
fig = plt.figure(figsize=(10, 8))

plt.subplot(131)
plt.plot([0, 1, 2], [3, 2, 5])

plt.subplot(132)
_ = plt.plot([0, 8, 12], [12, 2, 15], 'r--')

plt.subplot(133)
_ = plt.plot([0, 8, 12], [12, 2, 15], 'ro')
```

```python
_ = plt.bar(x=[0, 1, 2, 3], height=[1.2, 3.2, 1.2, 5.2])
```

```python
_ = plt.bar(x=['a', 'b', 'c', 'd'], height=[1.2, 3.2, 1.2, 5.2])
```

```python
_ = plt.barh(y=['a', 'b', 'c', 'd'], width=[1.2, 3.2, 1.2, 5.2])
```
